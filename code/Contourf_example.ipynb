```julia
using PyPlot
```

```julia
x = -10:0.01:10
y = -6:0.01:6
```

```julia
function Plot_func(x,y)
    x^4/10+(y)^4-6*x^2+(y-2)^2+0.5*x^3
end
```

```julia
z = [Plot_func(i,j) for i in x, j in y];
levels = -600:100:1000
```

```julia
fig,ax2 = subplots(constrained_layout=true)
CS = ax2.contourf(y,x,z,levels,cmap=plt.cm.bone)
fig.colorbar(CS)
savefig("/home/richard/tmp_image/Julia_exp/contourf_1")
show()
```
