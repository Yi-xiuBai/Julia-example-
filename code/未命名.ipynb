```julia
x = -10:0.01:10
function Aim_function(x)
    sin(x^3/40)+cos(x)
end
y = map(x->Aim_function(x),x);
```

```julia
plot(x,y)
```

```julia
delta = 0.1
x_0 = 6.7
resid =1e-6
δf = 1000
N=0
x1_set = []
while abs(δf) > resid && N<1e8
    δf = Aim_function(x_0 + delta)-Aim_function(x_0)
    x_0 = x_0 - δf * delta*10
    N = N+1
    x1_set = [x1_set;x_0]
    print(δf,x_0,"\n")
end
```
