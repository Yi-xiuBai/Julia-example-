## 1. plot 函数

```julia
x = collect(1:0.01:2)
y = rand(length(x));
```

```julia
xlabel("This is x label") #设置x轴标签
ylabel("This is y label") #设置y轴标签
grid() #添加格子
plot(x,y,linewidth = 4) #默认连线形式绘图，颜色默认为蓝色，线宽设置为4
plot(x,y,"o",c="r") #以圆圈形式绘图，颜色为红色
```

```julia
t = collect(0:0.2:5.0)
plot(t,t,"r--")
plot(t,t.^2,"bs")
plot(t,t.^3,"g^")
legend(["red desh line","blue square","green triangle"])
```

## 2.Scatter 函数

```julia
x = collect(0.0:2.0:50.0)
y = similar(x)
@. y = x^1.3+rand()*30
sizes = 500 .* rand(length(x)) .+ 200;
```

```julia
color_set = rand(length(x)); #设置每个marker为随机颜色
```

```julia
scatter(x,y,sizes,c=color_set, label="Luck",alpha = 0.5,marker=4) #alpha 参数可以调节透明度
```

### 自定义 marker的形状

```julia
theta = 0:0.01:2pi
f(x) = 1.6*cos(x),sin(x)^3
verts = map(f,theta); #通过给出N组 (x,y) 的点来勾画marker的形状
```

```julia
scatter(x,y,sizes,c=x,alpha = 0.6, label="Luck",marker=verts)
xlabel("δ")
ylabel("κ")
```
